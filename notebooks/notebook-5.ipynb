## Documentação

[Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html)

```python
import numpy as np
import pandas as pd
```

```python
dados = np.random.normal(size=(100, 10))

df = pd.DataFrame(dados, columns=[f'col{i+1}' for i in range(10)])
df.head()
```

```python
df.describe()
```

## Plots

[↑ Topo](#Pandas-III)

### DataFrame


```python
df.plot
```

```python
df.plot()
```

### Series

```python
df['col1'].plot()

# ou df.col1.plot()
```

### Plot por coluna

```python
df.loc[4].plot()
```

```python
df['g'] = df.apply(lambda x: 1 if x['col1'] - x['col2'] > 0.0 else 2, axis=1)
df[['col1', 'col2', 'g']].head()
```

```python
df2 = df[['col1', 'col2']]
df2.head()
```

## Tipos

[↑ Topo](#Pandas-III)

### Linha

- `kind=line`*
- `figsize=(8, 4)`*


- `DataFrame.plot.line`

*_Opcional_

```python
df2[['col1', 'col2']].plot(kind='line', figsize=(8, 4))
# ou df2.plot(kind='line', figsize=(8, 4))
```

```python
df2[['col1']].plot.line(figsize=(8, 4))
```

### Barra

[↑ Topo](#Pandas-III)

#### Vertical

```python
df2.head().plot.bar(figsize=(8, 4))
```

##### Stacked

```python
df2.head().plot.bar(stacked=True, figsize=(8, 4))
```

#### Horizontal

```python
df2.head().plot.barh(figsize=(5, 8))
```

### Histograma

[↑ Topo](#Pandas-III)

```python
df2.plot.hist()
```

```python
df2.plot.hist(alpha=0.5)
```

```python
df2.plot.hist(stacked=True)
```

```python
df2['col1'].plot.hist(orientation='horizontal', grid=True)
```

### BoxPlot

[↑ Topo](#Pandas-III)

```python
df.plot.box(figsize=(8, 6))
```

```python
df[df.columns[0:6]].plot.box(positions=[
    # Grupo 1
    1, 2, 3,
    
    # Grupo 2
    6, 7, 8
], figsize=(8, 6))
```

```python
df.groupby('g').plot.box()
```

### Área

[↑ Topo](#Pandas-III)

```python
# Por padrão é Stacked
df2[(df2['col1'] >= 0) & (df2['col2'] >= 0)].plot.area()
```

```python
df2[(df2['col1'] >= 0) & (df2['col2'] >= 0)].plot.area(stacked=False)
```

```python
df2.plot.area(stacked=False)
```

### Pontos

[↑ Topo](#Pandas-III)

```python
df2.head(30).plot.scatter(x='col1', y='col2', figsize=(8, 5))
```

```python
df.head(30).plot.scatter(x='col1', y='col2', c='col3', figsize=(8, 5))
```

```python
df.head(30).plot.scatter(
    x='col1',
    y='col2',
    c='g',
    colormap='viridis',
    figsize=(8, 5)
)
```

```python
df.head(30).plot.scatter(
    x='col1',
    y='col2',
    c='g',
    s=100,
    colormap='viridis',
    figsize=(8, 5)
)
```

```python
df.head(30).plot.scatter(
    x='col1',
    y='col2',
    s=df['col3'].head(30).abs()*200,
    figsize=(8, 5)
)
```
