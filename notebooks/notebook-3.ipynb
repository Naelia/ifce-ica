### Descrição das colunas
Apresenta:
- `count`: Contagem de ítens
- `mean`: Média dos valores
- `std`: Desvio padrão
- `min`: Menor valor
- `max`: Maior valor
- `25%`: Primeiro quartil
- `50%`: Segundo quartil ou Mediana
- `75%`: Terceiro quartil

````python
df.describe()
````

## Filtragem

### Linha

````python
# Linha de índice 4
df.loc[4]
````

````python
# Linhas 1, 3 e 5
df.loc[[1,3,5]]
````

````python
# Slice
df.loc[1:5]
````

````python
# Slice com step
df.loc[0:10:3]
````

````python
# Slice com step
df.loc[0::2] # Somente linhas pares
````
